```python
import math
import tqdm
import numpy as np
import pandas as pd

import pyarrow
import dask
import dask.array as da
import dask.bag as db
import dask.dataframe as dd
from dask.distributed import Client
# !pip install pyarrow==5.0 --user
```

```python
n = 234994989
FRACTION = 1e-3
# FRACTION = 200000/n
SAMPLE_PICKLE = "quotebank-sample.pickle"
DATASET_PATH = "../quotebank"
```

```python
client = Client()
client
```

```python
quotebank_df_array = [dd.read_parquet(f"{DATASET_PATH}/quotes-{year}.parquet", engine="pyarrow") for year in range(2008, 2021)]
dataset = dd.concat(quotebank_df_array, interleave_partitions=True)
dataset.head()
```

```python
n = 234994989
# !date
# with pbar:
#     n = len(dataset)
#     print(n == 234994989)
#     print(n)
# !date
```

```python
sample = dataset.sample(frac=FRACTION, replace=False).compute()
print(f"Sample length: {len(sample)}")
```

```python
import pickle as pickle
with open(SAMPLE_PICKLE, 'wb') as f:
    pickle.dump(sample, f, protocol=pickle.HIGHEST_PROTOCOL)
```
